Song Popularity Prediction

Given data about the top 50 spotify songs from 2019, let's try to predict whether a given song will be more or less popular.

We will be using five different models to make our predictions.

Explaination


NumPy and Pandas are used for numerical computations and data manipulation, respectively.

From scikit-learn (sklearn), various modules and classes are imported for preprocessing, model selection, and different classifiers (Logistic Regression, K-Nearest Neighbors, Decision Tree, MLP, SVM).

```python
import numpy as np
import pandas as pd
import plotly.express as px

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
```

Explanation:

A function onehot_encode is defined to perform one-hot encoding.
It takes DataFrame, column name, and prefix as input and returns the DataFrame with the specified column one-hot encoded.
Categorical columns 'Genre' and 'Artist.Name' are one-hot encoded using this function.

```python
def onehot_encode(df, column, prefix):
  df=df.copy()
  dummies=pd.get_dummies(df[column], prefix=prefix)
  df = pd.concat([df, dummies], axis=1)
  df=df.drop(column, axis=1)
  return df
```

```python
data = onehot_encode(data, 'Genre', 'genre')
data = onehot_encode(data, 'Artist.Name', 'artist')
```

```python
log_model = LogisticRegression()
knn_model = KNeighborsClassifier()
dec_model = DecisionTreeClassifier()
mlp_model = MLPClassifier()
svm_model = SVC()
```

```python
log_model.fit(X_train, y_train)
knn_model.fit(X_train, y_train)
dec_model.fit(X_train, y_train)
mlp_model.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
```

```python
log_acc = log_model.score(X_test, y_test)
knn_acc = knn_model.score(X_test, y_test)
dec_acc = dec_model.score(X_test, y_test)
mlp_acc = mlp_model.score(X_test, y_test)
svm_acc = svm_model.score(X_test, y_test)
```
